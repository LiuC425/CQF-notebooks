# Introduction

In this study, I analyze whether a Long Short-Term Memory model (LSTM) is able to predict a sign of daily logarithmic returns based on adjusted close prices. Thus, it is a classification problem, in which features are stock related time-series and target variable (labels) is the signum function of logarithmic returns (excluding observations with zeros).

Before an LSTM model is created, there are several imporant steps to do. First of all, I collect data that seem relevant and useful for the prediction of price movements. Besides open, high, low, close prices and volumes, we construct technical indicators (EMA, RSI, MACM, etc.), drift-independent volatility according to Yang & Zhang (2000) download and transform financial ratios, sentiment data and other data sources. In this study, we try to predict the daily stock price movements of Amazon (and verify the results separately with similar companies). We have two main data providers: Yahoo! Finance and Bloomberg.

After the data is collected, we apply several transformations to input data such that the model is able to capture miscellaneous relationships in our data. For example, we compute exponential moving average across various periods (5, 10, 100 days, etc.), take the difference between values of the same time-series over different windows (periods), calculate logarithmic changes and so on. In total, we obtain 266 features (we intentially generated that many ones).

Since many variables are generated, we need to select them carefully. We work with this problem as follows. Firstly, the data is inspected visually. I try to capture if there are relationships between generated features and the target variable. Secondly, a plain correlation analysis is performed. After that, we move towards more advanced techniques. In particular, we conduct the Principal Component Analysis (PCA) and ISOMAP analysis. Finally, we use an unsupervised machine learning technique - Self Organizing Maps (SOMs) to perform variable selection. Also, we take advantage of other methods such as Decision Tree feature importance and Gradient Boosting in Decision Trees to choose variables. When the analysis is complete, we refine the set of variables and construct one key dataset (SOM and DT based) and several additional datasets (PCA and ISOMAP).

When our datasets are ready, we can start designing models. It is decided to create two models: an LSTM model and a baseline model - logistic regression. The construction of an LSTM model can be summarized as follows. We prepare several architectures. After that a hyperparameter optimization process is conducted. When optimal parameters were found, we create a single model with optimal architecture and selected hyperparameters. Finally, the model is tested on a test data and key performance metrics are computed. For a baseline model we perform similar steps. In addition, we test the optimized model using PCA and ISOMAP datesets.

### Momentum Indicator

Let us define the **momentum** feature. This is nothing but the difference between a price at period *t* and a price *t-n*, where n is a look back period. <br> 
Also, we apply this function not only to prices but to other series as well.

```python
def momentum(data, out_series_name, 
             input_col, period=1, data_frame=False): 
    """
    This function computes the difference between a value at period t 
        and a value at period t-n.
    
    Parameters: 
        data: Pandas DataFrame object, which contains time-series in columns;
        out_series_name: string, required. This indicated the desired name of 
            an output series. 
            The name will have the following form : "mom_out_series_name_period";   
        input_col: string, requred. The name of a column, which is taken 
            to perform calculations. 
        period: integer, optional. Default value is 1. This is a look back period. 
        data_frame: binary, optional. Default False. This will define the output format.
    
    Returns:
        If the data_frame flag is set True, the function returns data, 
            but with an added column. 
            This column contains the computed series with name out_series_name. 
        Otherwise, the function returns Pandas Series contains the computed 
            series with name out_series_name. 
    """
    
    selected_series = data[input_col].copy()
    mom = selected_series - selected_series.shift(period)
    if(data_frame == True):
        out_col_name= 'mom_%s_%d' % (out_series_name, period)
        data[out_col_name] = mom
        return(data)
    else:
        return(mom)
```

### Logarithmic Return <a name="Log_Return"></a>

The function below computes the logarithmic difference between two values (primarily, close prices) at period *t* and *t-n*, where n is a look back period.

```python
def log_ret(data, out_series_name, 
            input_col, period=1, data_frame=False):
    
    """
    This function computes the logarithmic difference between a value 
        at period t and a value at period t-n.
    
    Parameters: 
        data: Pandas DataFrame object, which contains time-series in columns;
        out_series_name: string, required. This indicated the desired name of 
            an output series. 
            The name will have the following form : "mom_out_series_name_period";   
        input_col: string, requred. The name of a column, which is taken 
            to perform calculations. 
        period: integer, optional. Default value is 1. This is a look back period. 
        data_frame: binary, optional. Default False. This will define the output format.
    
    Returns:
        If the data_frame flag is set True, the function returns data, 
            but with an added column. 
            This column contains the computed series with name out_series_name. 
        Otherwise, the function returns Pandas Series contains the computed series 
            with name out_series_name. 
    """
    
    series = data[input_col].copy()
    log_ret = np.log(series / series.shift(period))
    if(data_frame == True):
        out_col_name= 'log_ret_%s_%d' % (out_series_name, period)
        data[out_col_name] = log_ret
        return(data)
    else:
        return(log_ret)
```

### Exponential Moving Average (EMA)

The exponential moving average is computed using formulas 1 and 2.

$$ \textit {EMA}_{t}(p)=\alpha \cdot p_{t}+(1-\alpha )\cdot \textit  {EMA}_{t-1}, {\quad} (1)$$

where the coefficient $\alpha$ represents the degree of weighting decrease and defined as:

$$\ \alpha ={\frac  {2}{n+1}}, {\quad} (2)$$

 We do not use the `.ewm()` method embeded in `pandas` bacause it uses slighly different definition. We would like to disregard the look back period in the output. For example, for 50 days EMA, we expect to get 49 NaN values in the output EMA series. This interesting aspect is well-described in this [article](https://towardsdatascience.com/trading-toolbox-02-wma-ema-62c22205e2a9).

```python
def ema(data, out_series_name, input_col, 
        period=21, data_frame=False):
    """
    This function computes Exponential Moving Average (EMA) of
        values (close prices) over a given period.
    
    Parameters: 
        data: Pandas DataFrame object, required;
        out_series_name: string, required. This indicated the desired name of 
            an output series. 
            The name will have the following form : "ema_%out_series_name_%period";  
        period: integer, optional. Default 21. Used as a period (number of days), 
            over which EMA is computed;
        input_col: string, requred. The name of a column, which contains close prices;
        data_frame: binary, optional. Default False.
   Returns:
        If the data_frame flag is set True, the function returns data, 
            but with an added column. 
            This column contains the computed series with name out_series_name. 
        Otherwise, the function returns Pandas Series contains the computed series 
            with name out_series_name. 
    """
    # make a copy of prices to avoid unnecessary changes in input
    close_price = data[input_col].copy()
    # compute sma over a required period
    modified_close_price = close_price.rolling(period).mean()
    # compute ema over a required period
    ema = modified_close_price.ewm(span=period, adjust=False).mean() 
    
    if(data_frame == True):
        out_col_name = 'ema_%s_%d' % (out_series_name, period)
        # concatenate the computed series to the initial data frame`
        data[out_col_name] = ema
        return(data)
    else:
        return(ema)
```

### Relative Strenght Index (RSI) <a name="RSI"></a>

The RSI is classified as a momentum oscillator, measuring the velocity and magnitude of directional price movements. Usually, it is bounded between 0 and 100. The RSI provides investors with signals whether a stock is overbought or oversold. We want an LSTM model to use this information as well. If it works, the model should capture this. The method is well-described in this [article](https://en.wikipedia.org/wiki/Relative_strength_index). To compute it, we compute upward `U` and downard `D` changes. `U` measures are charaterized by the situation when the close being higher than the previous close. If close prices do not change over a period `U` and `D` are zeros. The `RSI` it itself is computed as in formula 5 using smoothed moving average. Also, the smoothed moving averages should be appropriately initialized with a simple moving average using the first `N` values in the price series. The detailed information about the RSI is provided in the `rsi` function below.

$$ U = close_{t} - close_{t-1},  {\quad} (3)$$
$$ D = 0 $$


$$ U = 0 $$
$$ D = close_{t-1} - close_{t},  {\quad} (4)$$

$$ RSI = 100*{ EMA_{N}(p) \; of \; U \over EMA_{N}(p) \; of \; U + EMA_{N}(p) \; of \; D },  {\quad} (5) $$

```python
def rsi(data, out_series_name, period=14, 
        input_col='Adj Close', data_frame = False):
    """
    This function computes Relative Strenght Index (RSI) over a given period (window).
    
    Parameters: 
        data: Pandas DataFrame object, required. The dataframe must contain close prices;
        out_series_name: string, required. This indicated the desired name of an output 
        series. 
            The name will have the following form : "rsi_%out_series_name_%period";
        period: integer, optional. Default 14. Used as a period (number of days), 
            over which RSI is computed;
        input_col: string, required. The name of a column, which contains close prices;
        data_frame: binary, optional. Default False.
   Returns:
        If the data_frame flag is True, the function returns the DataFrame that is used as an argument,
        but with added column, which contains the computed values. 
        Otherwise, the function returns Pandas Series. 
    """ 
    
    close_price = data[input_col].copy() # compute close price
    # compute the difference between current and previous prices
    diff_now_prev = close_price - close_price.shift() 
    # comput ups
    up = diff_now_prev.apply(lambda x: x if x > 0. else 0)
    # compute the difference between previous and current prices
    diff_prev_now = close_price.shift() - close_price
    # comput downs
    down = diff_prev_now.apply(lambda x: x if x > 0. else 0)
    # comput smoothed moving averages 
    smma_up = wwma(up, period)
    smma_down = wwma(down, period)
    # comput rs 
    rs = smma_up/smma_down
    # comput rsi such that it lies in range [0, 100]
    rsi = 100 - 100/(1+rs)
    
    if(data_frame == True): # condition on the output
        
        col_name = 'rsi_%s_%d' % (out_series_name, period)
        data[col_name] = rsi
        return(data)
    
    else:
        return(rsi)
```

###  Stochastic Oscillator

Again, the stochastic oscillator is a momentum indicator that uses resistance and support levels. In this project, we compute only the so-called `stochastic K oscillator`. The formula 6 and the function `fast_stochastic` clearly illustrate how to compute the oscillator. The indicator was designed by Dr. George Lane and the additional information about it can be glimpsed at this [article](https://en.wikipedia.org/wiki/Stochastic_oscillator).

$${\displaystyle \%K=100*\frac{(close_{t}-low_{N})}{(high_{N}-low_{N})}},  {\quad} (6) $$

```python
def fast_stochastic(data, out_series_name, high_col='High', low_col='Low', 
                    close_col='Adj Close', period=14, data_frame = False):
    """
    This function computes the Stochastic K indicator over a given period.
    
    Parameters: 
        data: Pandas DataFrame object, required. The dataframe must contain close, high, low prices;
        out_series_name: string, required. This indicated the desired name of an output series. 
            The name will have the following form : "s_k_%out_series_name_%period";
        period: integer, optional. Default 14;
        close_col: string, required. Default 'Close'. The name of a column, which contains close prices;
        high_col: string, required. Default 'High'. The name of a column, which contains high prices;
        low_col: string, required. Default 'Low'. The name of a column, which contains low prices;
        data_frame: binary, optional. Default False.
   Returns:
        If the data_frame flag is True, the function returns the DataFrame that is used as an argument,
        but with added column, which contains the computed values. 
        Otherwise, the function returns Pandas Series.
    """ 
    
    # extract close, high and low prices from data
    data_copy = data[[close_col, high_col, low_col]].copy()
    # compute the lowest price over a period
    low_min = data_copy[low_col].rolling(period).min()
    # compute the highest price over a period
    high_max =  data_copy[high_col].rolling(period).max()
    # compute stochastic oscillator as in formula 6
    k_fast = 100 * (data_copy[close_col] - low_min)/(high_max - low_min)
    
    if(data_frame == True):
        
        col_name_s_k = 's_k_%s_%d' % (out_series_name, period)
        col_name_s_d = 's_%s_d_%d' % (out_series_name, period)
        data[col_name_s_k] = k_fast
        return(data)
    else:
        return(k_fast)
```

### Moving Average Convergence Divergence (MACD) <a name="MACD"></a>

The MACD indicator is a very simple calculation in comparison with previous ones. Basically, it is the difference between EMA over period M and EMA over the period N. M is usually 12 and N is 26. In this project, apart from 12 and 26 days, we use other windows.

$$ MACD = EMA(M) - EMA(N),  {\quad} (6) $$

```python
def macd(data, out_series_name, close_col='Adj Close', 
         period_long_ema=26, period_short_ema= 12, 
         data_frame = False):
    """
    This function computes Moving Average Convergence Divergence using (adjusted close)
    prices over a given period (window).
    
    Parameters: 
        data: Pandas DataFrame object, required;
        out_series_name: string, required. This indicated the desired name of an output series. 
            The name will have the following form : "MACD_%out_series_name_%period";
        period_long_ema: integer, optional. Default 26;
        period_short_ema : integer, optional. Default 12;
        close_col: string, Required. Default 'Adj Close'. The name of a column, which contains (adjusted) close prices;
        data_frame: binary, optional. Default False.
   Returns:
        If the data_frame flag is True, the function returns the DataFrame that is used as an argument,
        but with added column, which contains the computed values. 
        Otherwise, the function returns Pandas Series.
    """ 
      
    long_ema = ema(data, out_series_name = out_series_name, period=period_long_ema, 
                   input_col='Close', data_frame=False)
    short_ema = ema(data, out_series_name = out_series_name, period=period_short_ema, 
                    input_col='Close', data_frame=False)
    MACD = long_ema - short_ema
    
    if(data_frame == True):
        col_name = 'MACD_%s_%d_%d' % (out_series_name, period_long_ema, period_short_ema)
        data[col_name] = MACD
        return(data)
    else:
        return(MACD)
```

### Commodity Channel Index (CCI) <a name="CCI"></a>

The CCI index is created to track cyclical movements in commodity and other asset prices. Formulas 7 and 8 clearly illustrate how CCI is computed. First of all, one computes the so-called *typical price (TP)* as in formula 7. Then, one computes the difference between *TP* and SMA of prices. The difference is divided by MAD, which stands for *mean absolute deviation*. The detailed interpretation of the CCI can be found in this [article](https://en.wikipedia.org/wiki/Commodity_channel_index). The detailed description about CCI construction can be found in the comments of function `cci`

$$ {TP_{t} = p_{t} = {\frac {high_{t} + low_{t} + close_{t}}{3}} },  {\quad} (7)  $$


$$ {\displaystyle CCI_{t}={\frac {1}{0.015}}\cdot 
    {\frac {p_{t}-{\textit {SMA}}_{n}(p_{t})}{{\textit {MAD}}_{n}(p_{t})}}},  {\quad} (8)  $$

```python
def cci(data, out_series_name, period=10, close_col = 'Adj Close', 
        high_col = 'High', low_col = 'Low', data_frame = False): 
    """
    This function computes Commodity Channel Index (CCI) over a given period (window) from close, high, 
    and low prices.
    
    Parameters: 
        data: Pandas DataFrame object, required. The dataframe must contain close, high, low prices;
        out_series_name: string, required. This indicated the desired name of an output series. 
            The name will have the following form : "CCI_%out_series_name_%period";
        period: integer, required. Default 10. Used as a period (number of days), over which CCI is computed;
        close_col: string, required. Default 'Adj Close'. The name of a column, which contains close prices;
        high_col: string, required. Default 'High'. The name of a column, which contains high prices;
        low_col: string, required. Default 'Low'. The name of a column, which contains low prices;
        data_frame: binary, optional. Default False.
   Returns:
        If the data_frame flag is True, the function returns the DataFrame that is used as an argument,
        but with added column, which contains the computed values. 
        Otherwise, the function returns Pandas Series. 
    """  
    # make a copy of prices to avoid unnecessary changes in input
    data_copy = data.copy() 
    # compute typical price as in formula 7
    tp = (data_copy[high_col] + data_copy[low_col] + data_copy[close_col]) / 3 
    # compute cci according to the given definition
    cci = (tp - tp.rolling(period).mean()) \
                    / (0.015 * tp.rolling(period).std()) 
    
    if(data_frame == True):
        # create a name for a column to add
        col_name = 'CCI_%s_%d' % (out_series_name, period)
        # concatenate the computed series to the initial data frame
        data[col_name] = cci 
        
        return(data)
    else:
        
        return(cci)
```

### Average True Range (ATR) <a name="ATR"></a>

The average true range in indicator unlike previously considered ones is volatility measure. It does not provide an investor with the idea of a price direction but with the idea of price volatility. Despite the fact that it might look that the indicator is hard to compute, it is actually very simple. Basically, it tracks the difference between high and low prices in TR (computed in formula 9), and then it is computed as exponential moving average (or modified moving average), which we create in the function `wwma`. The complete procedure and the interpretation to this indicator can be found in this [article](https://en.wikipedia.org/wiki/Average_true_range). Also, all construction steps are well-documented in function `atr` and `wwma`.

$${\displaystyle {\text{TR}}={\max({\text{high}},{\text{close}}_{\text{t-1}})-\operatorname {min} ({\text{low}},{\text{close}}_{\text{t-1}})}\,  {\quad} } (9)$$

$$ ATR_{t}={{ATR_{{t-1}}\times (n-1)+TR_{t}} \over n} $$

```python
def wwma(series, period):
    """
    This function computes smoothed moving average (SMMA) over a given period.
    
    Parameters: 
        series: Pandas Series, DataFrame, required;
        period: integer, optional. Default 10. Used as a period (number of days), over which SMMA is computed;
    Returns:
        The function returns Pandas Series.
    """
    
    return(series.ewm(alpha=1/period, adjust=False, min_periods=period).mean())
```

```python
def atr(data, out_series_name, period=14, close_col = 'Adj Close', 
        high_col = 'High', low_col = 'Low', data_frame = False):
    """
    This function computes Average True Range (ATR) over a given period (window) using close, high, low prices.
    
    Parameters: 
        data: Pandas DataFrame object, required. The dataframe must contain close, high, low prices;
        out_series_name: string, required. This indicated the desired name of an output series. 
            The name will have the following form : "atr_%out_series_name_%period";
        period: integer, required. Default 14. Used as a period (number of days), over which ATR is computed;
        close_col: string, required. Default 'Close'. The name of a column, which contains close prices;
        high_col: string, required. Default 'High'. The name of a column, which contains high prices;
        low_col: string, required. Default 'Low'. The name of a column, which contains low prices;
        data_frame: binary, optional. Default False.
   Returns:
        If the data_frame flag is True, the function returns the DataFrame that is used as an argument,
        but with added column, which contains the computed values. 
        Otherwise, the function returns Pandas Series. 
    """ 
    # extract close, high and low prices from data
    data_copy = data[[close_col, high_col, low_col]].copy()
    # compute absolute difference between currect high and current low
    data_copy['tr0'] = abs(data_copy[high_col] - data_copy[low_col])
    # compute absolute difference between currect high and previous close 
    data_copy['tr1'] = abs(data_copy[high_col] - data_copy[close_col].shift())
    # compute absolute difference between currect low and previous close 
    data_copy['tr2'] = abs(data_copy[low_col] - data_copy[close_col].shift())
    tr = data_copy[['tr0', 'tr1', 'tr2']].max(axis=1)
    atr = wwma(tr, period)
    if(data_frame == True):
        col_name = 'atr_%s_%d' % (out_series_name, period)
        data[col_name] = atr
        return(data)
    else:
        return(atr)
```

### Bollinger Bands (BB)

Finally, we construct one of the most promising measures in this section. The Bollinger Bands indicator characterizes the price dynamics. Basically, it produces two bands: upper and lower. These bands are constructed based on simple moving averages and standard deviation times the number of standard deviations. It requires two parameters: `N` (periods over which one should compute SMA from close prices) and the number of standard deviations to indicate the width of the BB indicator. The interpretation and detailed instruction to construct the index can be found in this [article] (https://en.wikipedia.org/wiki/Bollinger_Bands). In addition, one can observe all construction steps in the `bollinger_bands` function below.

$$ \text{Upper Bollinger Band} = SMA(close, N) + \text{number of standard deviations}*st.dev(close, N) $$
$$ \text{Lower Bollinger Band} = SMA(close, N) - \text{number of standard deviations}*st.dev(close, N) $$

```python
def bolinger_bands(data, out_series_name, input_col='Adj Close', 
                   period=20, num_of_std = 2, data_frame = False):
    """
    This function computes Bolinger Bands over a given period (window).
    
    Parameters: 
        data: Pandas DataFrame object, required;
        period: integer, optional. Default 20.
        num_of_std: integer, optional. Default 2. Number of standart deviations
        close_col: string, required. Default 'Close'. The name of a column, which contains close prices;
        data_frame: binary, optional. Default False.
   Returns:
        If data_frame is True, the function returns the DataFrame that is used as an argument 
        but with 2 added column, which contain the computed values
        for upper and lower bands. Otherwise, the function returns two Pandas Series.
    """
    
    price = data[input_col].copy()
    rolling_mean = price.rolling(window=period).mean()
    rolling_std  = price.rolling(window=period).std()
    
    upper_band = rolling_mean + (rolling_std*num_of_std)
    lower_band = rolling_mean - (rolling_std*num_of_std)
    
    if(data_frame == True):
        col_name_upper = 'upper_BB_%s_%d' % (out_series_name, period)
        col_name_lower = 'lower_BB_%s_%d' % (out_series_name, period)
        data[col_name_upper] = upper_band
        data[col_name_lower] = lower_band
        return(data)
    else:
        return((upper_band, lower_band))
```

## Drift-Independent Volatility Estimator <a name="Drift_Vol"></a>

Drift-independent volatility estimatorn is know as very powerful estimator to measure volatility of stock prices. It was proposed by Dennis Yang and Qiang Zhang its construction is illustrated in their [paper](http://www.atmif.com/papers/range.pdf). Unlike other volatility measures, it takes into account
an overnight jump. The step-by-step procedure is well-documented below in the `drift_independent_vol_estimator` function. The function was found on the internet, however, it was tested and modified to fit the project.

```python
import math # package for math computations

def drift_independent_vol_estimator(data, out_series_name, window=21, 
                  trading_periods=252, clean=False, data_frame=False):
    """
    This function computes drift-independent volatility as in Yang & Zhang (2000).
        
    Parameters: 
        data: Pandas DataFrame object, required;
        out_series_name: string, required. This indicated the desired name of an output series. 
            The name will have the following form : "vol_YZ_%out_series_name_%period";
        window: integer, optional. Default value is 21.
        trading periods: integer, optional. Default value is 252.
        clean: boolean, optional. Default value is False. It controls wether the function
            drops NaNs or not.
        data_frame: boolean, optional. Default values is False.
    
    Returns:a
        If data_frame is True, the function returns the DataFrame that is used as an argument 
        but with 2 added column, which contain the computed values
        for upper and lower bands. Otherwise, the function returns two Pandas Series.
    """
    
    # upload price data
    price_data = data[['Adj Close', 
                       'Open', 'Close', 
                       'High', 'Low']].copy()
    
    # compute logarithmic difference between high and open prices
    log_high = (price_data['High'] / price_data['Open']).apply(np.log)
    # compute logarithmic difference between low and open prices
    log_low = (price_data['Low'] / price_data['Open']).apply(np.log)
    # compute logarithmic difference between close and open prices
    log_close_open = (price_data['Adj Close'] / price_data['Open']).apply(np.log)
    # compute logarithmic difference between open and  lagged close prices
    log_open_close = (price_data['Open'] / price_data['Adj Close'].shift(1)).apply(np.log)
    log_open_close_sq = log_open_close**2
    # compute logarithmic return between close and lagged close
    log_close_close = (price_data['Adj Close'] / price_data['Adj Close'].shift(1)).apply(np.log)
    log_close_close_sq = log_close_close**2
    # compute the proposed measure
    measure = log_high * (log_high - log_close_open) + log_low * (log_low - log_close_open)
    # compute the close based vol
    close_vol = log_close_close_sq.rolling(
        window=window, center=False).sum() * (1.0 / (window - 1.0))
    # compute the open based vol
    open_vol = log_open_close_sq.rolling(window=window, 
                          center=False).sum() * (1.0 / (window - 1.0))
    
    window_measure = measure.rolling(window=window,
                              center=False).sum() * (1.0 / (window - 1.0))
    # compute k as in paper
    k = 0.34 / (1 + (window + 1) / (window - 1))
    
    output = (open_vol + k *close_vol + (1 - k * window_measure)).apply(np.sqrt)* math.sqrt(trading_periods)
    
    if(data_frame == True):
        col_name = 'vol_YZ_%s_%d' % (out_series_name, window)
        if(clean == True):
            data[col_name] = output.dropna()
            return(data)
        else:
            data[col_name] = output
            return(data)
    else:
        if(clean == True):
            return(output.dropna())
        else:
            return(output)
```

Let us split the data into 3 sub-samples: *train/development/test*. Our features were generated and stored in a single `pandas` data frame called `aggregate_data`. I created a separate function (below) for this task in order to add more flexibility and increase the robustness of this study. The additional data set is created for **cross-validation** purposes. It is very important to prevent any information leakage during cross-validation** and testing processes.

Let us define the function for splitting the data into 3 pieces.

```python
def train_dev_test_split(df, dev_share = 0.10, test_share = 0.10):
    
    """ This function splits the data and performs scaling if required.
        The split has the following sequential structure: [train, development, test].

    Parameters: 
        df: Pandas DataFrame object, which features and labels in columns;
        dev_share:, float, optional. Default value is 0.10. The share of the data from the tail
            before the test sample that will be cut for development purposes (before the test set).
        test_share:, float, optional. Default value is 0.10. The share of the data from the tail 
            to be taken for testing.
   Returns:
        Train, development and tests samples (pandas data frames).
    """
    
    train_share = 1.0 - dev_share - test_share
    df_train = df.iloc[0:int(train_share*df.shape[0]),:]
    df_dev = df.iloc[int(train_share*df.shape[0]):int((train_share+dev_share)*df.shape[0]),:]
    df_test = df.iloc[int((train_share+dev_share)*df.shape[0]):,:]
        
    return(df_train, df_dev, df_test)
```

Now, we split our data in the following proportion: 
* train: 60%;
* development: 25%;
* test: 15%.

We allocate a quarter of the data to the development sample since it is important for us to have enough data for optimization. However, during the study, I have tried various combinations to balance all three samples (to have an equal share of positive classes in each sub-sample). However, I did not make a big difference in the output and conclusion.

```python
train_data, dev_data, test_data = train_dev_test_split(aggregate_data.iloc[:,1:], 
                                                       dev_share = 0.25, test_share = 0.15)
```

Now, let us **scale** the data. The scaling process is important because our features might differ from each other (in absolute terms) quite a lot (e.g. prices and returns). Scaling increases chances to converge to the optimum (faster).

By using `MinMaxScaler` estimator from` sklearn.preprocessing`. `MinMaxScaler` we scale our data such that each feature lies between zero and one (mapping to $ [0, 1] $).

One can also try to use `StandardScaler`. `StandardScaler` standardizes features by removing the mean and scaling to unit variance. Empirically, I found that other algorithms such as **Self Orginizing Maps (SOM)** work better with `MinMaxScaler` (for SOM, all else equal, *quantization* and *topological* errors are lower).

However, in scaling process, it is quite important to preven any **informational leakage**. In fact, it is a common mistake when a researcher scales a training set and a validation set separately. This immediately casts doubt a validation process because we need to apply the same scaling factors for a validation set as we derived for a training set.

```python
from sklearn.preprocessing import scale, MinMaxScaler # Import MinMaxScaler
# from sklearn.preprocessing import StandardScaler 

scaler = MinMaxScaler() # Introduce scaler

train_data_scaled = scaler.fit_transform(train_data) # scale training data
dev_data_scaled = scaler.transform(dev_data) # transform a validation set using transformation applied to training 
test_data_scaled = scaler.transform(test_data) # transform a test set using transformation applied to training

# transform numpy arrays to pandas dataframes
train_data_scaled_df = pd.DataFrame(scaler.fit_transform(train_data), columns = train_data.columns)
dev_data_scaled_df = pd.DataFrame(scaler.transform(dev_data), columns = train_data.columns)
test_data_scaled_df = pd.DataFrame(scaler.transform(test_data), columns = train_data.columns)

X_train_scaled = train_data_scaled[:,:-1] # select features by exluding labels 
y_train_scaled = train_data_scaled[:,-1] # select labels by exluding features 

X_dev_scaled = dev_data_scaled[:,:-1] # select features by exluding labels 
y_dev_scaled = dev_data_scaled[:,-1] # select labels by exluding features 

X_test_scaled = test_data_scaled[:,:-1] # select features by exluding labels 
y_test_scaled = test_data_scaled[:,-1] # select labels by exluding features

# let save the data in the directory to access it from other applications (R)
# pd.DataFrame(train_data_scaled, columns = aggregate_data.columns.tolist()[1:]).to_csv('train_data_scaled.csv')
```

## Principal Component Analysis (PCA) <a name="Visual_Inspection"></a>

After we conducted a visual inspection, performed a correlation analysis, we understand that there are variables, which are pretty similar. For our problem, we would like to reduce the set of features. The PCA method is very powerful for this. The PCA yields an orthogonal basis in which different individual dimensions of the data are uncorrelated.

```python
# import PCA from sklearn
from sklearn.decomposition import PCA 
# let us set up 50 components to construct
pca = PCA(n_components=30)

# apply the same idea, which we used for scaling
# we use obtain components from training dataset
# and obtain PCs for dev and test samples accordingly

X_train_pca = pca.fit_transform(X_train_scaled)
X_dev_pca = pca.transform(X_dev_scaled)
X_test_pca = pca.transform(X_test_scaled)

# extract the percentage of variance explained
share_explained = round(pd.DataFrame(np.cumsum(pca.explained_variance_ratio_)*100),1)
share_explained.rename(columns = {0: '% of Explained Varience'}, inplace = True)
```

### Self-Orgnizing Maps: Feature Selection <a name="Visual_Inspection"></a>

The self-organizing-map is a powerful unsupervised machine learning method. The method maps a multidimensional space of training examples to a low-dimensional (e.g. two dimensional) space. Thus, the procedure solves the problem of dimensionality reduction. The [paper](https://link.springer.com/article/10.1007/BF00337288) was introduced by Teuvo Kohonen in 1982. The algorithm works as follows.

1. Initialize the weights of the neurons (or coordinates) randomly or based on PCA to speed the learning process up;
2. Then, a number of example vectors is uploaded;
3. After a training example is uploaded, its Euclidean distance to all weights vectors is computed;
4. Then, we find the Best Matching Unit (BMU) as the neuron (unit) with the weights vector that is the closest to the input (in terms of Euclidean distance).
5. Then, iteratively, the weights in the grid are adjusted towards the input vector.

The formula below formalizes this process.

$$ W_{v}(i+1)=W_{v}(i)+\theta (i,u,v)\cdot \alpha (i)\cdot (D(t)-W_{v}(i)), $$

where i is the step-index, t an index into the training sample, u is the index of the BMU for the input vector D(t), $\alpha(s)$ is a monotonically decreasing learning coefficient; $\theta (u, v, i)$ is the neighborhood function which gives the distance between the neuron u and the neuron v in step i. More information can be found in [article](http://www.scholarpedia.org/article/Kohonen_network). Now, let us discuss the practical aspects of the process.

Below, we perform the following steps to select features:
1. select the size of a grid;
2. select the topology of a grid;
3. initialize weights (in our case, we use PCA);
4. initialize hyperparameters;
5. train SOM;
6. analyze quantization and topographic errors;
7. optimize hyperparameters such that we minimize quantization and topographic errors;
8. analyze SOMs of several variable;
9. perform feature selection using the `som_feature_selection` function;
10. compare the selecte by SOM features with ones obtained using decision tree based methods;
11. combine relevant features in a refined set;
12. analyze SOMs of the features from the refined set of variable.
13. create a final dataset.m

aAs far as I know, there is no strictly defined procedure for the selection of grid size. We take into account the following recommendations from practitioners. First of all, we set up a grid size proposed by Vesano (Advances in Computational Intelligence: 13th International Work-Conference, 2015). Also, we take advantage of the `kohonen` package available in `R` to see if grid size is not too sparse. We require that one cell possess at least 10 observations. Also, we try to achieve homogeneity (in terms of the number of observations per cell) in the structure of a grid.
We come up to the size of 15. Later, we initialize weights based on PCA and set up some guessed hyperparameters (learning rate, sigma, and topology). After we got our estimates, we train SOM and obtain quantization and topographic errors. Then, we iteratively optimize SOM to achieve minimal errors.

After the SOM is trained, we compute SOMs (weights) and look at the maps. To select the most relevant features, we try to find such SOMs of explanatory variables that would mimic the SOM of our target variables. On the other hand, we can exclude variables with identical SOMs. Since it is a relatively burdening procedure, I tried to think and search for other methods for feature selection based on SOM. It appears that the method illustrated in this [paper](https://link.springer.com/chapter/10.1007%2F978-3-319-89935-0_6) provides an interesting scheme to select variables (the main idea is still similar to the one we explained above). The algorithm of the method can be summarized as follows:

* create N-1 random features;
* combine them with the existent variables;
* features are normalized (mapped to $[0,1]$);
* initialize a hyperparameter `a`
* create two check matrices:
* abs(vertically stacked arrays constructed from the target feature minus normalized weights matrix);
* abs(vertically stacked arrays constructed from the target feature plus normalized weights matrix minus one);
* perform logical OR operation: (check matrix <= 1 or check matrix <= 1) and assign the output to a matrix S;
* in matrix S, select a column which has a target index and assign zero;
* perform row-wise summation of the matrix S and assign the output to a variable S2
* finally, select the features with the highest sum and remove those ones which are less or equal to zero.

When we completed the procedure, we again plot SOMs of the selected features and compare them with a target variable to ensure the correctness of the method.

Finally, we add features obtained from the feature selection procedure based on decision trees and again plot SOMs.
If we do not find very similar ones and the added features reflect some patterns from the SOM of a target variable, 
we say that the dataset is refined and can be used further for model development.

```python
from minisom import MiniSom

# np.sqrt(5*np.sqrt(train_data_scaled.shape[0]))
size = 15
som = MiniSom(size, size, train_data_scaled.shape[1],
              neighborhood_function='gaussian', sigma=3,
              learning_rate = 0.6, topology = 'rectangular', 
              random_seed=1)
# initializes the weights to span the first two principal components
som.pca_weights_init(train_data_scaled)
```

```python
def som_feature_selection(Weights, feature_names, 
                          target_index = -1, 
                          a = 0.04):
    """ This function performs feature selection based on the computed matrix by
            using Self Orginizing Maps
    Parameters: 
        Weights: numpu array, required. This parameter stands for the weight matrix computed
            using SOM procedure (MiniSom package).
        feature_names: list, required. This parameter contains the names of the features preserving
            the order as in Weights.
        target_index: integer, required. Default value is -1. This parameter contains idx of 
            the target feature (labels).
        a: float, required. This is a hyperparameter. Approximate range of values [0.03, 0.06].
    Returns:
        selected_features: list of strings. It cointains the names 
            of the selected features in order of selection
        target_name: string. The name of the target variable.
    """
    #reshapes Weights into MxN assuming M neurons and N features
    Weights_2d_array = np.reshape(Weights, (Weights.shape[0]*Weights.shape[1], 
                          Weights.shape[2])) 
    
    target_name = feature_names[target_index]
    # create N -1 random features
    Rand_feat = np.random.uniform(low=0, high=1, 
                                  size=(Weights_2d_array.shape[0], 
                                        Weights_2d_array.shape[1] - 1)) 
    # add them to the N regular ones
    W_with_rand = np.concatenate((Weights_2d_array, 
                                  Rand_feat), axis=1) 
    # normalize each feature between 0 and 1
    W_normed = (W_with_rand - W_with_rand.min(0)) / W_with_rand.ptp(0) 
    
    # select a column with a target feature
    Target_feat = W_normed[:,target_index] 

    # Two conditions to check against a
    
    Check_matrix1 = abs(np.vstack(Target_feat) - W_normed)
    Check_matrix2 = abs(np.vstack(Target_feat) + W_normed - 1)
    
    # applies "or" element-wise in two matrices
    
    S = np.logical_or(Check_matrix1 <= a, 
                      Check_matrix2 <= a).astype(int) 

    S[:,target_index] = 0 #ignore the target feature so that it is not picked

    selected_features = []
    while True:
        
    # add all rows for each column (feature)
        S2 = np.sum(S, axis=0) 

        if not np.any(S2 > 0): # if all features add to 0 kill
            break
            
        # feature with the highest sum gets selected first
        selected_feature_index = np.argmax(S2) 
        
        # if random feature is selected kill
        if selected_feature_index > (S.shape[1] - (Rand_feat.shape[1] + 1)): 
            break

        selected_features.append(feature_names[selected_feature_index])

        # delete all rows where selected feature evaluates to 1,
        # thus avoid selecting complementary features
        rows_to_delete = np.where(S[:,selected_feature_index] == 1)
        S[rows_to_delete, :] = 0
        
        
    return selected_features, target_name
```

## Long Short-Term Memory Model: Data Preparation <a name="#LSTM_initialization"></a>

In order to set up an LSTM model in `Keras`, first of all, we should prepare the data. 
It is known that LSTM models are very useful in analyzing time-series since they can preserve some important information and delete useless for the analysis data. Thus, we need to set up a so-called *look back* period for our data. This will define how far the model will look back.

Let us transform the refined dataset from section 5.5.4 obtained using SOM and Decision Tree-based methods.
The refined dataset has two dimensions. The number of rows shows the number of examples, the number of columns defines the number of features used plus a target variable.

If we want to create a data set such that an LSTM model created in `keras` is able to process it, we need to transform input data into three-dimensional `numpy` array. The reshaped data set will have the following shape: **(number of examples, look back period, number of features)**. In order to this, let us define a function.

```python
def transform_for_LSTM(df, look_back_period):
    """
    This function transforms two dimensional pandas 
        dataframe or numpy array to three dimensional numpy array.
    The input data should contain features in columns.
    For each feature, it creates additional shifted columns. 
    Each column is shifted by 1 period. The number
    of columns for a feature will be equal to look_back_period.
    
    Parameters: 
        df: Pandas DataFrame object or two dimensional ndarray, requred.
        look_back_period: int, required. 
    Returns:
        3-dimensional numpy ndarray. The shape of the returned ndarray is defined as df.shape[0], look_back_period, df.shape[1].
    """
    
    nrFeatures = df.shape[1] # define number of features in df
    nrDays = df.shape[0] # define number of examples in df
    look_back = look_back_period # define desired look back period

    X = [] # empty list to store ndarrays (2 dimensional) with shape = (nrDays, look_back)
    y = [] # empty list to store ndarrays (2 dimensional) with shape = (nrDays, 1)
    
    if (type(df)  == np.ndarray): # check the typep of input data

        for j in range(0, nrFeatures):
            x_feature = [] # empty list to store a single feature with pre-defined look back (shifted)
            for i in range(look_back, nrDays):
                x_feature.append(df[i-look_back:i,j]) # create a shifted feature and store in x_feature
            X.append(np.array(x_feature)) # append ndarray to X

        for i in range(look_back, nrDays): # check the typep of input data
                y.append(df[i,-1])

        X_array, y_array = np.array(X), np.array(y) # transform lists to ndarrays
        X_lstm = X_array.reshape(X_array.shape[1], 
                                 X_array.shape[2], X_array.shape[0]) # redefine shape
        y_lstm = y_array # reassign labels
                       
        # the code below performes exactly the same manipulations but with pandas dataframes    
            
    else:
                       
        for j in range(0, nrFeatures):

            x_feature = []
            for i in range(look_back, nrDays):
                x_feature.append(df.iloc[i-look_back:i,j])
            X.append(np.array(x_feature))

        for i in range(look_back, nrDays):
                y.append(df.iloc[i,-1])

        X_array, y_array = np.array(X), np.array(y)
        X_lstm = X_array.reshape(X_array.shape[1], X_array.shape[2], 
                                                   X_array.shape[0])
        y_lstm = y_array

    return(X_lstm, y_lstm)
```

We define a *look back* period as **30** days. Apparently, this is a hyperparameter, which is subject to optimization. During the project, several periods were tested, but I decided to stick with 30 days in this report.

## Long Short-Term Memory Model: Setup & Hyperparameters Search  <a name="#LSTM setup"></a>

The LSTM model was developed in 1997 by two bright scientists Sepp Hochreiter and Jürgen Schmidhuber.
Their paper [Hochreiter, Schmidhuber (1997)](https://www.researchgate.net/publication/13853244_Long_Short-term_Memory) had a wonderful impact on sequence modeling. This paper provides details not only about the model itself but also discusses the theory of **vanishing gradients** (the main problem that GRU and LSTM models address).
Thus, we recommend reading the paper to understand all aspects of the model. Nevertheless, I provide key equations to summarize the model.

The model is comprised of a few elements. The first element is a *memory cell* $\tilde{C}_{t}$.

$$\tilde{C}_{t} = \tanh(W_{c}[a_{t-1}, x_{t}]+b_{c})$$

Then, we have an *updated gate* $\Gamma_{u}$. The *gate* can be derived asm

$$ \Gamma_{u} = W_{u}[a_{t-1}, x_{t}]+b_{u}) $$

If we compare LSTM with GRU, LSTM has not only an *update* unit, but also a *forget* unit. This *forget* unit is calculated as

$$\Gamma_{f}= \sigma(W_{f}[a_{t-1}, x_{t}]+b_{f})$$

Also, we have an *output* gate $\Gamma_{o}$, which is computed as

$$\Gamma_{o}= \sigma(W_{o}[a_{t-1}, x_{t}]+b_{o})$$

And then, the update value for the memory cell is $C_{t}$, which equals to

$$ C_{t} = \Gamma_{u} \circ \tilde{C}_{t} +  \Gamma_{f} \circ  \tilde{C}_{t-1}$$

Finally, $a_{t}$ is equal to

$$ a_{t} = \Gamma_{o} \circ \tanh(C_{t})$$

Thus, we summarized the equations governing the behavior of LSTM.

After the dictionary with hyperparameters is created, we need to create a model. Our model possess the architecture 1. The model is created by means of the `lstm_model_1` function.

```python
def lstm_model_1(x_train, y_train, x_val, y_val, params):
    
    """
    This function creates an archuture of an LSTM model such that
        talos is able to search over hyperparameters.
    Parameters: 
        x_train: numpy 3-dimensional array, requred. The shape of x_train must be
             defined (number of examples, look_back_period, number of features).
        y_train: numpy 1-dimensional array, required. Target variable.
        
        x_val: numpy 3-dimensional array, requred. This is a validation sample.
        The shape of x_train must be
             defined (number of examples, look_back_period, number of features).
        y_val: numpy 1-dimensional array, required. Target variable.
        
    Returns:
        history, model: talos objects.
    """
    
    model = Sequential() # initiate sequential model
    
    # set up the first LSTM layer
    model.add(LSTM(units=params['lstm_layer_1_units'],
                   return_sequences=True))
    # set up a dropout layer associated with the first LSTM layer
    model.add(Dropout(params['drop_out']))
    
    # set up the second LSTM layer
    model.add(LSTM(units=params['lstm_layer_2_units'], return_sequences=False))
    # set up a dropout layer associated with the first LSTM layer
    model.add(Dropout(params['drop_out']))

    # set up a dense layer
    model.add(Dense(params['dense_layer_1_units'],
                    activation = params['dense_layer_1_act']))
    
    # set up the output layer
    model.add(Dense((1), activation='sigmoid'))
    
    # set up the model: loss fn, optimization scheme, metric to optimize
    model.compile(loss='binary_crossentropy', 
                  optimizer = params['optimizer'](lr=lr_normalizer(params['lr'], 
                                                                   params['optimizer'])), 
                  metrics=['accuracy'])
    
    history = model.fit(x=x_train, # train sample with features
                        y=y_train, # train sample with labels
                        validation_data=[x_val, y_val], # validation sample with features
                        epochs = params['epochs'], # validation sample with labels
                        batch_size = params['batch_size'], # batch size
                        verbose=0)

    return(history, model)
```

Set up a selected architecture as a function. It is already commented above and it is very straightforwards.

```python
def LSTM_optimal(optimal_parameters_LSTM):
    
    # replace the hyperparameter inputs with references to params dictionary 
    
    model = Sequential()
    
    model.add(LSTM(units=optimal_parameters_LSTM['lstm_layer_1_units'],
                   activation='tanh',
                   return_sequences=True))
    
    model.add(Dropout(optimal_parameters_LSTM['drop_out']))

    model.add(LSTM(units=optimal_parameters_LSTM['lstm_layer_2_units'],
                   activation='tanh',
                   return_sequences=False))
    
    model.add(Dropout(optimal_parameters_LSTM['drop_out']))
    
    model.add(Dense(optimal_parameters_LSTM['dense_layer_1_units'], 
                    activation=optimal_parameters_LSTM['dense_layer_1_act']))
    
    model.add(Dense((optimal_parameters_LSTM['dense_layer_output_units']), 
                    activation=optimal_parameters_LSTM['dense_layer_output_act']))
    
    return(model)
```

Let us define a simple function to plot a confusion matrix.

```python
def plot_confusion(y_test, y_pred, classes):
    
    confusion_matrix_new = confusion_matrix(y_test, y_pred)
    normalized_cm = confusion_matrix_new \
                    * 1. / confusion_matrix_new.\
                        sum(axis=1)[:, np.newaxis]  # normalize confusion matrix
    plt.imshow(normalized_cm, 
               interpolation='nearest', 
               cmap=plt.cm.Blues)
    plt.colorbar()
    ticks = np.arange(len(classes))
    plt.xticks(ticks, 
               classes, rotation=45)
    plt.yticks(ticks, classes)
    plt.xlabel('Predicted Classes')
    plt.ylabel('Actual Classes')
    plt.title('Confusion Matrix ')
    plt.tight_layout()
    plt.show()
```

Let us introduce a very simple function to plot a score curve that ensures that we found optimal regulariztion parameter for the l1 and l2 regularized logistic classifiers.

```python
def plot_scores(optimizer):
    scores=[]
    for i in range(len(optimizer.cv_results_['params'])):
        scores.append([optimizer.cv_results_['params'][i]['C'], 
                optimizer.cv_results_['mean_test_score'][i],
                optimizer.cv_results_['std_test_score'][i]])
    scores = np.array(scores)
    plt.semilogx(scores[:,0], scores[:,1])
    plt.fill_between(scores[:,0], scores[:,1]-scores[:,2], 
                                  scores[:,1]+scores[:,2], alpha=0.3)
    plt.title('Grid search to find the optimal parameter C (inverse of regularization strength)')
    plt.xlabel('Hyperparameter C')
    plt.ylabel('Score')
    plt.show()
```

```python
def plot_roc_curve(model, X_test, y_test):
    
    y_proba = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _=metrics.roc_curve(y_test,  y_proba)
    auc = metrics.roc_auc_score(y_test, 
                                y_proba)
    plt.plot(fpr,tpr,label = "AUC=" + str(round(auc,4)))
    plt.legend(loc=2)
    plt.plot([0, 1], [0, 1], linestyle='--') # plot no skill
    plt.title('ROC-Curve')
    plt.show()
```
